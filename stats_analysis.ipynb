```python
from src.stats_on_game import play_multiple_game
from src.player import Player, SmartPlayer, RandomTrickPlayer
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Create players
players = [SmartPlayer("Alice"), SmartPlayer("Bob"), SmartPlayer("Charlie"),  SmartPlayer("Sara"),
           RandomTrickPlayer("David"), RandomTrickPlayer("Eve"), RandomTrickPlayer("Frank"), RandomTrickPlayer("Grace")]
```

```python
def plot_game_stats(game_stats_df):
    plt.figure(figsize=(20, 6))
    sns.lineplot(data=game_stats_df)
    plt.show()
```

```python
## SANS LE PETIT POISSON

game_stats_df = play_multiple_game(players=players, n_games=1000, with_poisson=False)
mean_per_10_games = game_stats_df.groupby(game_stats_df.index // 10).mean() 
plot_game_stats(mean_per_10_games)
print(game_stats_df.mean())
print(game_stats_df.std())
```

```python
## AVEC LE PETIT POISSON

game_stats_df = play_multiple_game(players=players, n_games=1000, with_poisson=True)
mean_per_10_games = game_stats_df.groupby(game_stats_df.index // 10).mean() 
plot_game_stats(mean_per_10_games)
print(game_stats_df.mean())
print(game_stats_df.std())
```
